# VISÃO COMPUTACIONAL - TRATAMENTO DE IMAGENS COM Pillow

OBJETIVO:
- Realizar práticas de visão computacional aplicadas a imagens utilizando recursos da biblioteca Pillow.

Para abrir uma imagem fora do Jupyter:
- from PIL import Image
- imagem = Image.open("tokyoTemple.jpg")
- imagem.show(): abre a imagem fora do Jupyet Notebook

```python
#ABRINDO IMAGEM DENTRO DO JUPYTER:

from IPython.display import display
imagem = Image.open('C:/Users/sylvi/tokyoTemple.jpg')
display(imagem)
```

```python
#Descobrindo dimensões de altura e largura da imagem:
largura, altura = imagem.size
print('Largura', largura)
print('Altura', altura)
```

```python
#Salvando uma imagem:
imagem.save('C:/Users/sylvi/cópiaTokyoTemple.jpg')
```

```python
#Definindo retângulo para cortar uma imagem:
posições = (150,150,300,300) #dimensões do retângulo: esquerda, cima, direita, baixo
imagemCortada = imagem.crop(posições)
display(imagemCortada)
```

```python
#Colocando imagem eme scala de cinza:
imagemCinza = imagem.convert('L')  #L=escala de cinza
display(imagemCinza)
```

```python
#Convertendo imagem de volta para RGB:
imagemRGB = imagem.convert('P', palette=Image.ADAPTIVE, colors=256)  #ADAPTIVE = converte preservando alguns elementos da imagem
display(imagemRGB)                                                   #Neste caso, as cores foram preservadas.
```

```python
#Rotacionando a imagem:
rotacionada = imagem.rotate(180)  #ângulo de rotação deve ser informado dentro dos parênteses
display(rotacionada)
```

```python
#Redimensionando a imagem:
redimensionada = imagem.resize((400,200))
display(redimensionada)
```

```python
#Ajustando o brilho da imagem:
from PIL import ImageEnhance
brilho = ImageEnhance.Brightness(imagem)
novaImagem = brilho.enhance(1.5)  #define o nível do brilho
display(novaImagem)
```

```python
#Ajustando o contraste da imagem:
from PIL import ImageEnhance
contraste = ImageEnhance.Contrast(imagem)
novaImagem2 = contraste.enhance(2.1)  #define o nível do brilho
display(novaImagem2)
```

```python
#Coletando um pixel específico da imagem:
pixel = imagem.getpixel((100,70))
```

```python
print(pixel)  #Retorna o RGB específico do pixel pesquisado
```
